```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df=df.drop(columns=["test preparation course"])
df
```

```python
df.info()
```

```python
sns.histplot(x=df["math score"], kde=True)
```

```python
df["math score"].describe
```

```python
x=df["writing score"]
```

# math_score ni normallashtirish

```python
df["Normalizition"]=(x-x.mean())/x.std()
df.head()
```

# Normallashtirishdan oldin va keyingi grafiklari

```python
fig , ax=plt.subplots(1,2 , figsize=(12,5))
sns.histplot(df["writing score"],ax=ax[0],kde=True)
sns.histplot(df["Normalizition"],ax=ax[1],kde=True,color="y")
```

# Writing score ni 0 va 1 oraligiga olib kelish yani - ( min-max normalization)

```python
df["Min_Max"]=(x-x.min())/(x.max()-x.min())
df.head()
```

```python
fig , ax=plt.subplots(1,2 , figsize=(12,5))
sns.histplot(df["writing score"],ax=ax[0],kde=True)
sns.histplot(df["Min_Max"],ax=ax[1],kde=True,color="y")
```

```python
df["Quantile"]=(x-x.median())/(x.quantile(0.75)-x.quantile(0.25))
df.head()
```

```python
fig , ax=plt.subplots(1,2 , figsize=(12,5))
sns.histplot(df["writing score"],ax=ax[0],kde=True)
sns.histplot(df["Quantile"],ax=ax[1],kde=True,color="y")
```

Hammasi birida

```python
fig , ax=plt.subplots(3,2 , figsize=(12,20))

sns.histplot(df["writing score"],ax=ax[0,0],kde=True)
sns.histplot(df["Normalizition"],ax=ax[0,1],kde=True,color="y")

sns.histplot(df["writing score"],ax=ax[1,0],kde=True)
sns.histplot(df["Min_Max"],ax=ax[1,1],kde=True,color="y")

sns.histplot(df["writing score"],ax=ax[2,0],kde=True)
sns.histplot(df["Quantile"],ax=ax[2,1],kde=True,color="y")
```
