```python
import numpy as np
```

```python
# Ikki matritsani ko'paytirish
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

product = np.dot(matrix1, matrix2)
print("Matritsa ko'paytirish:\n", product)

# Determinantni hisoblash
det = np.linalg.det(matrix1)
print("Determinant:", det)

# Invers matritsa
inv_matrix = np.linalg.inv(matrix1)
print("Invers matritsa:\n", inv_matrix)
```
