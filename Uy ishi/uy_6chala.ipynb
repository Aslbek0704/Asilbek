```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("Sales.csv" , index_col=0)
df
```

# Qayerlarga koproq sotiladi ? (Bar va Pie korinishlarda)

```python
Regionls=df["Region"].unique()
count=df["Region"].value_counts()

fig , ax=plt.subplots()
ax.bar(Regionls, count, label="Region")
ax.set_title('Bar diagramma')
plt.legend()
plt.show()
```

```python
Regionls=df["Region"].unique()
count=df["Region"].value_counts()

fig , ax=plt.subplots()
ax.pie(count, labels=Regionls, explode=(0,0.1,0), autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Pie diagramma')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV faylni yuklab olish
df = pd.read_csv('data.csv')

# 2022 yili uchun o'rtacha qiymatlarni hisoblash
df['Average 2022'] = df[['Jan 2022', 'Feb 2022', 'Mar 2022', 'Apr 2022', 'May 2022', 'Jun 2022',
                         'Jul 2022', 'Aug 2022', 'Sep 2022', 'Oct 2022', 'Nov 2022', 'Dec 2022']].mean(axis=1)

# Grafikni tayyorlash
plt.figure(figsize=(10, 6))
plt.bar(df['State'], df['Average 2022'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Average Value for 2022')
plt.title('Average Values for 2022 by State')
plt.xticks(rotation=45)
plt.tight_layout()

# Grafikni ko'rsatish
plt.show()
```
