```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.read_csv("Tuition.csv",index_col=0)
df.head()
```

# Degree boyicha br chiziq boylab xarajatlarini chiqarish jitter vazifasi nuqtalarni ustma ust tuwrmaydi

```python
sns.stripplot(data=df,
              y="Degree",
              x="Cost",
              jitter=True,
              hue="Degree")
```

# Swarmplot da korinishi

```python
sns.swarmplot(data=df,
              y="Degree",
              x="Cost",
              hue="Degree")
```

# Boxplotda korinishi

```python
sns.boxplot(data=df,
            x="Cost",
            y="Degree",
            hue="Degree")
```

# Violinplotda korinishi

```python
sns.violinplot(data=df,
            x="Cost",
            y="Degree",
            hue="Degree")
```
