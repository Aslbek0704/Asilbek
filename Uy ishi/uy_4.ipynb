```python
import pandas as pd

data = {
    'Ism': ['Ali', 'Vali', 'Salim', 'Oisha', 'Aziz', 'Bobur', 'Jamila', 'Anvar'],
    'Yosh': [25, 30, 22, 27, 24, 35, 28, 26],
    'Shahar': ['Toshkent', 'Samarqand', 'Buxoro', 'Toshkent', 'Buxoro', 'Toshkent', 'Samarqand', 'Toshkent'],
    'Oylik': [1200000, 1500000, 900000, 1300000, 1100000, 1800000, 1400000, 1200000],
    'Kasb': ['Dasturchi', 'Menejer', 'Dizayner', 'Dasturchi', 'Muhandis', 'Menejer', 'Dizayner', 'Muhandis'],
    'Tajriba': [3, 5, 2, 4, 3, 8, 6, 4]
}

df = pd.DataFrame(data)
kasb_oylik_tajriba = df.groupby('Kasb').agg({'Oylik': 'mean', 'Tajriba': 'mean'}).reset_index()
print(kasb_oylik_tajriba)
```
