```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("train.csv")
df.head()
```

```python
sns.catplot(x="Embarked",
            y="Age",
            data=df,
            hue="Sex",
            kind="point"
)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.catplot(x="Embarked",
            y="Age",
            data=df,
            hue="Sex",
            kind="point"
)
plt.show()
```

```python
sns.set_style("ticks")
sns.catplot(x="Embarked",
            y="Age",
            data=df,
            hue="Sex",
            kind="point"
)
plt.show()
```

```python
sns.set_style("dark")
sns.catplot(x="Embarked",
            y="Age",
            data=df,
            hue="Sex",
            kind="point"
)
plt.show()
```

```python
sns.set_style("ticks")
sns.catplot( x="Pclass",
             data=df,
             kind="count",
             )
plt.show()
```

```python
sns.set_palette("RdBu_r")
sns.catplot( x="Pclass",
             data=df,
             kind="count",)
plt.show()
```

```python
sns.set_context("notebook")
sns.catplot( x="Pclass",
             data=df,
             kind="count",)
plt.show()
```

```python
g=sns.catplot(x="Pclass",
            data=df,
            kind="box",
            y="Fare")
g.figure.suptitle("New Title" , y=1.1)
```

```python
sns.catplot(x="Pclass",
            data=df,
            kind="box",
            y="Fare",
            col="Survived"
            )
```

```python
g=sns.catplot(x="Pclass",
            data=df,
            kind="box",
            y="Fare",
            col="Survived")
g.figure.suptitle("New Title" , y=1.1)
```

```python
g=sns.catplot(x="Pclass",
            data=df,
            kind="box",
            y="Fare",
            col="Survived",
            hue="Pclass")
g.figure.suptitle("New Title" , y=1.1)


g.set(xlabel="New X Label",  
ylabel="New Y Label") 
```

```python
g=sns.catplot(x="Pclass",
            data=df,
            kind="box",
            y="Fare",
            hue="Pclass")
g.figure.suptitle("New Title" , y=1.1)
plt.xticks(rotation=90)
```
